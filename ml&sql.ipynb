```python
import sqlite3

# Connect to the same SQLite database file
conn = sqlite3.connect("project_db.db")
cursor = conn.cursor()

# Query the table
cursor.execute("SELECT label, path FROM images")
rows = cursor.fetchall()  # Each row is (label, path)

# Print each row
for row in rows:
    print(row)

# Close the connection
cursor.close()
conn.close()
```

```python
# === Step 2: Define the augmentation function ===
def augment_image(img):
    """Returns a list of augmented versions of a PIL image."""
    augmented = []

    # Original
    augmented.append(img)

    # Horizontal Flip
    augmented.append(ImageOps.mirror(img))

    # Vertical Flip
    augmented.append(ImageOps.flip(img))

    # Rotations
    augmented.append(img.rotate(15))
    augmented.append(img.rotate(-15))

    # Brightness
    brightness = ImageEnhance.Brightness(img)
    augmented.append(brightness.enhance(1.5))  # Brighter
    augmented.append(brightness.enhance(0.7))  # Darker

    # Contrast
    contrast = ImageEnhance.Contrast(img)
    augmented.append(contrast.enhance(1.4))
    augmented.append(contrast.enhance(0.6))

    return augmented
```

```python
# --- Step 5: Load, resize, and augment images ---
import os
from PIL import Image
import numpy as np

image_data = []
labels = []

# Your augment_image function must be defined elsewhere
# Example: def augment_image(img): return [img]

for label, path in rows:
    if os.path.exists(path):
        try:
            img = Image.open(path).convert('RGB').resize((128, 128))
            augmented_images = augment_image(img)  # your augmentation function

            for aug_img in augmented_images:
                img_array = np.array(aug_img)
                image_data.append(img_array)
                labels.append(label)  # only label, no color

        except Exception as e:
            print(f"Error loading {path}: {e}")
    else:
        print(f"File not found: {path}")

# --- Step 6: Done ---
print(f"Total images after augmentation: {len(image_data)}")
```
